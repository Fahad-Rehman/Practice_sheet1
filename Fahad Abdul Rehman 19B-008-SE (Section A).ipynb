Q5: A football is kicked 1 meter from the ground at an angle of 30 degree with a velocity of 15 m/sec. 
<br>Calculate the following:
<br>a. Max height above the ground.
<br>b. Total time in the air.
<br>c. Total distance traveled.
<br>d. Velocity vector at max height.
<br>e. Velocity when the ball hits the ground.

![proj](proj.jpg)



```python
from math import pi
from math import sin
from math import cos
from math import sqrt
angle = 30
radians = float(angle * pi/180)
velocity = 15
height = 1
g = 9.8 #Acceleration due to gravity

#Part a: Maximum height of football`
H =  height + ((velocity**2) * (sin(radians)**2)) / (2 * g)
print("The total height of the football is: " + str(H) + " m")

#Part b: Total time of flight
T = (velocity * sin(radians) + sqrt((velocity**2) * (sin(radians)**2) + (2 * g * height))) / g
print("The total time of flight of the football is: " + str(T) + " s")

#Part c: Total distance travelled
R = (velocity * cos(radians)) * (velocity * sin(radians) + sqrt((velocity**2) * (sin(radians)**2) + (2 * g * height))) / g
print("The total distance travelled by the football is: " + str(R) + " m")

#Part d: Velocity vector at max height
v_maxheight = velocity * H
print("The velocity vector at max height is: " + str(v_maxheight))

#Part e: Velocity when ball hits the ground
velocity_final = H * -g
print("The velocity with which the ball hits the ground is: " + str(velocity_final) + " m/s")
```

Q7: Three resiters of 10, 20 and 30 ohms are connected in series and in parallel if 10 V is given to them find current and voltages across them.

```python
V = 10
r1 = 10
r2 = 20
r3 = 30

#For series connection, current remains same
R_series = r1 + r2 + r3
print("The total resistance in series connection is: " + str(R_series) + " ohms")
I_series = (V / R_series)
print("The total current in series connection is: " + str(I_series) + " A")
V1 = (I_series * r1)
print("The voltage across first resistor in series connection is: " + str(V1) + " V")
V2 = (I_series * r2)
print("The voltage across second resistor in series connection is: " + str(V2) + " V")
V3 = (I_series * r3)
print("The voltage across third resistor in series connection is: " + str(V3) + " V")

#For parallel connection, voltage remains same
print("The total voltage in parallel connection is: " + str(V) + " V")
I_1_parallel = (V / r1)
print("The current across first resistor in parallel connection is: " + str(I_1_parallel) + " A")
I_2_parallel = (V / r2)
print("The current across second resistor in parallel connection is: " + str(I_2_parallel) + " A")
I_3_parallel = (V / r3)
print("The current across third resistor in parallel connection is: " + str(I_3_parallel) + " A")
```
